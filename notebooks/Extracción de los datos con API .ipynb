# Extracción de datos con API
Se va a realizar la extracción de los datos con la API proporcionada por la plataforma Apify para extraer los comentarios de 5 vídeos de TikTok relacionados con la donación de sangre. Todos los videos los han realizado cuentas oficiales españolas y creadores de contenido españoles para intentar centralizar de la mejor manera posible el estudio a España.

#### Video 1: Nachter

```python
from apify_client import ApifyClient
import pandas as pd
import json

from dotenv import load_dotenv
import os

load_dotenv()
APIFY_TOKEN = os.getenv("APIFY_TOKEN")

video_links = [
    "https://www.tiktok.com/@nachter/video/7357337111676079392?_r=1&_t=ZN-93r9JxE2rN0",
    # añadir más aquí si funciona
]


client = ApifyClient(APIFY_TOKEN)


run = client.actor("clockworks/tiktok-comments-scraper").call(
    run_input={
        "postURLs": video_links,
        "maxComments": 800,
        "maxRepliesPerComment": 40
    }
)

print("Run ID:", run["id"])
print("Dataset ID:", run["defaultDatasetId"])
```

```python
## Obtención de resultados
items = list(client.dataset(run["defaultDatasetId"]).iterate_items())

print("Total comentarios descargados:", len(items))

# -----------------------------
# Guardar en JSON
# -----------------------------
with open("nachter_tiktok.json", "w", encoding="utf-8") as f:
    json.dump(items, f, ensure_ascii=False, indent=4)

# -----------------------------
# Guardar en CSV
# -----------------------------
df = pd.DataFrame(items)
df.to_csv("nachter_tiktok.csv", index=False)

print("Archivos guardados correctamente")
```
